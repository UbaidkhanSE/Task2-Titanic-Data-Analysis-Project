```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Data Cleaning
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)
titanic.dropna(subset=['embark_town', 'deck'], inplace=True)
titanic['survived'] = titanic['survived'].astype(str)  # Convert to string for categorical plotting

# Basic Information about the dataset
print("Basic Information about the Titanic Dataset:")
print(titanic.info())
print("\nSummary Statistics of the Titanic Dataset:")
print(titanic.describe(include='all'))

# Exploratory Data Analysis (EDA)

# 1. Survival Rate by Passenger Class
plt.figure(figsize=(10, 6))
sns.barplot(x='pclass', y=titanic['survived'].astype(int), data=titanic, errorbar=None, palette='Blues')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# 2. Survival Rate by Gender
plt.figure(figsize=(10, 6))
sns.barplot(x='sex', y=titanic['survived'].astype(int), data=titanic, errorbar=None, palette='Set2')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

# 3. Survival Distribution by Age
plt.figure(figsize=(10, 6))
sns.histplot(x='age', data=titanic, hue='survived', bins=30, kde=True, multiple='stack', palette='Set1')
plt.title('Survival Distribution by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# 4. Survival Count by Embarkation Port
plt.figure(figsize=(10, 6))
sns.countplot(x='embarked', hue='survived', data=titanic, palette='Set1')
plt.title('Survival Count by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Count')
plt.show()

# 5. Fare Distribution
plt.figure(figsize=(10, 6))
sns.histplot(x='fare', data=titanic, bins=30, kde=True, color='green')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()

# 6. Survival Rate by Number of Siblings/Spouses Aboard (SibSp)
plt.figure(figsize=(10, 6))
sns.barplot(x='sibsp', y=titanic['survived'].astype(int), data=titanic, errorbar=None, palette='muted')
plt.title('Survival Rate by Number of Siblings/Spouses Aboard')
plt.xlabel('Number of Siblings/Spouses Aboard')
plt.ylabel('Survival Rate')
plt.show()

# 7. Survival Rate by Number of Parents/Children Aboard (Parch)
plt.figure(figsize=(10, 6))
sns.barplot(x='parch', y=titanic['survived'].astype(int), data=titanic, errorbar=None, palette='pastel')
plt.title('Survival Rate by Number of Parents/Children Aboard')
plt.xlabel('Number of Parents/Children Aboard')
plt.ylabel('Survival Rate')
plt.show()

# 8. Pairplot to see relationships between multiple variables
sns.pairplot(titanic[['age', 'fare', 'pclass', 'survived']].astype({'survived': 'category'}), hue='survived', palette='husl')
plt.show()

# 9. Correlation Heatmap to see correlation between numerical features
plt.figure(figsize=(12, 8))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()

# 10. Survival Rate by Deck
plt.figure(figsize=(10, 6))
sns.barplot(x='deck', y=titanic['survived'].astype(int), data=titanic, errorbar=None, palette='cubehelix')
plt.title('Survival Rate by Deck')
plt.xlabel('Deck')
plt.ylabel('Survival Rate')
plt.show()

# 11. Age Distribution by Passenger Class and Gender
plt.figure(figsize=(14, 8))
sns.boxplot(x='pclass', y='age', hue='sex', data=titanic, palette='Set3')
plt.title('Age Distribution by Passenger Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()

# 12. Survival Rate by Ticket Class and Embarkation Port
plt.figure(figsize=(10, 6))
sns.barplot(x='pclass', y=titanic['survived'].astype(int), hue='embarked', data=titanic, errorbar=None, palette='viridis')
plt.title('Survival Rate by Passenger Class and Embarkation Port')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```
